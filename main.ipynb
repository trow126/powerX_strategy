```python
from datetime import datetime
import pandas as pd
import talib

```

```python
df_rates = pd.read_pickle('./df_rates.pkl')
df_rates
```

```python
df_rates['rsi'] = talib.RSI(df_rates['close'], timeperiod=7)
df_rates['stoch_k'], df_rates['stoch_d'] = talib.STOCH(df_rates['high'], df_rates['low'], df_rates['close'],
                                                       fastk_period=14, slowk_period=3, slowd_period=3)
df_rates['macd'], df_rates['macd_signal'], _ = talib.MACD(df_rates['close'],
                                                          fastperiod=12, slowperiod=26, signalperiod=9)

df_rates
```
